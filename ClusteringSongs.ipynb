# Song classes of various emotions

Importing and preprocessing

```python
import pandas as pd
df=pd.read_csv("./sample_data/songs-bv3.csv",index_col=0)
df.head(10)
```

```python
df.info()
df = df[df.columns[3:14]]
```

```python
df=df.drop(['artist','popularity','liveness','speechiness'], axis=1)
df.info()
```

Simple scatterplots for feature comparisons

```python
import plotly.express as px
fig = px.scatter_matrix(df.head(50),
width=1200, height=1600)
fig.show()
```

Performing K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import plotly.graph_objects as go
import numpy as np
X=df
scaler = MinMaxScaler()
scaler.fit(X)
X=scaler.transform(X)
inertia = []
for i in range(1,11):
    kmeans = KMeans(
        n_clusters=i, init="k-means++",
        n_init=10,
        tol=1e-04, random_state=42
    )
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
fig = go.Figure(data=go.Scatter(x=np.arange(1,11),y=inertia))
fig.update_layout(title="Inertia vs Cluster Number",xaxis=dict(range=[0,11],title="Cluster Number"),
                  yaxis={'title':'Inertia'})
```

```python
kmeans = KMeans(
        n_clusters=5, init="k-means++",
        n_init=50,
        tol=1e-04, random_state=25
    )
kmeans.fit(X)
clusters=pd.DataFrame(X,columns=df.columns)
clusters['label']=kmeans.labels_
polar=clusters.groupby("label").mean().reset_index()
polar=pd.melt(polar,id_vars=["label"])
fig4 = px.line_polar(polar, r="value", theta="variable", color="label", line_close=True,height=800,width=1400)
fig4.show()
```

```python
pie=clusters.groupby('label').size().reset_index()
pie.columns=['label','value']
px.pie(pie,values='value',names='label')
```

```python

clusters= clusters.drop(['danceability','energy','valence','acousticness','instrumentalness','tempo','loudness'], axis=1)
clusters.head()
```

```python
df=pd.read_csv("./sample_data/songs-bv3.csv",index_col=0)
df = pd.concat([df, clusters], axis=1)
```

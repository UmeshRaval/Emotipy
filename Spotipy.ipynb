Getting india hindi language only category playlist and getting tracks

```python
results = sp.categories(country='IN',locale='HI_IN',limit =50)
categories = results['categories']['items']
cat_id=[]
for category in categories:
  cat_id.append(category['id'])

playlists = []
for id in cat_id:
  try:
    results = sp.category_playlists(category_id=id, country='IN', limit=50, offset=0)
  except:
    print('error token')
  items = results['playlists']['items']
  for item in items:
    try:
      playlists.append(item['uri'])
    except:
      print('error item')
print(playlists)
```

Getting the tracks and printing track number in playlist

```python
tracks = []
counter = len(playlists)
for playlist in playlists:
  print(counter)
  counter-=1;
  results = sp.playlist_tracks(playlist, fields=None, limit=100, offset=0, market=None, additional_types=('track', ))
  items = results['items']
  for item in items:
    try:
      tracks.append(item['track']['uri'])
    except:
      print('err track')
print(tracks)
```

```python
print(len(tracks))
tracks_ver = list(set(tracks))
print(len(tracks_ver))
```

Using pandas to store the dataset and extracting it into CSV format

danceability: danceability measures whether the track is suitable for dancing (0: least danceable, 1: most danceable)

energy: energy measures the intensity and activity of a track, less energy score means the track is less energetic

valence: valence measures the positiveness from a track, more valence score means the track is sound positive (happy, joy, cheerful)

acousticness: acousticness measures whether the track is acoustic, 1 represent high confidence the track is acoustic

instrumentalness: instrumentalness measures the likelihood of a track containing vocals, a high score means the track is likely instrumental

tempo: the overall estimated tempo of a track in beats per minute (BPM)

loudness: the overall loudness of a track in decibels (dB)

```python
import pandas as pd
rows_list = []
tracks = tracks_ver
count = len(tracks)
for track in tracks:
    try:
      print(count)
      count-=1
      result = sp.track(track)
      url = result['external_urls']['spotify']
      uri = track
      name  = result['name']
      popularity = result['popularity']
      artist = result['artists'][0]['name']
      features = sp.audio_features(track)
      danceability = features[0]['danceability']
      energy = features[0]['energy']
      valence = features[0]['valence']
      acousticness = features[0]['acousticness']
      instrumentalness = features[0]['instrumentalness']
      tempo = features[0]['tempo']
      loudness = features[0]['loudness']
      speechiness = features[0]['speechiness']
      liveness = features[0]['liveness']

      dict1  = dict(name=name,url=url,uri=uri,popularity=popularity,artist = artist,danceability=danceability,energy=energy,valence = valence,acousticness = acousticness,instrumentalness = instrumentalness,tempo = tempo,loudness = loudness,speechiness=speechiness,liveness =liveness)
      rows_list.append(dict1)
    except:
      print('error track')

df = pd.DataFrame(rows_list)
```
